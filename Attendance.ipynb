```python
import random
import networkx as nx
import matplotlib.pyplot as plt
from ABM_model import AttendanceModel
```

```python
test_G = nx.gnm_random_graph(10, 20)
for u, v, d in test_G.edges(data=True):
    d['weight'] = random.uniform(0.1, 5)
```

```python
test_model = AttendanceModel(test_G.number_of_nodes(), test_G)
test_model.run_model()
```

```python
test_agent_res = test_model.datacollector.get_agent_vars_dataframe()
test_model_res = test_model.datacollector.get_model_vars_dataframe()
```
