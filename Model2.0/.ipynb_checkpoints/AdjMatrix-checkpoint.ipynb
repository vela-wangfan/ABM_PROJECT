# AdjMatrix_initialization 

### Adjacency Matrix initialization: 
This adjacency matrix should represent the initial network of relationship within the class: 

* The k degree of each agent i is sampled from a one-parameter discrete poisson distribution

* For each agent i, k agents j are sampled from a uniform distribution
* The weight of the relationship is then set with a uniform distribution between 0 and 1 

For a completely disconnected network is enough to set k=0. 
In the following examples you can visualize the initial ajdacency matrix for different k values in order to choose the one that you see fit for your simulations
